2.1. Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()
data
```

Диаграмма рассеяния (scatter plot): позволяет визуализировать взаимосвязь между площадью и стоимостью.

```python
df.plot.scatter(x='sqft_lot', y='price')
plt.show()
```

Диаграмма рассеяния между количеством спален и стоимостью

```python
df.plot.scatter(x='bedrooms', y='price')
plt.show()
```

Столбчатая диаграмма (bar plot): позволяет сравнить значения категориальной переменной.


```python
df.groupby('grade')['price'].mean().plot.bar()
plt.show()
```

```python
df.boxplot(by='bathrooms', column='price')
plt.show()
```

Ящик с усами (boxplot): позволяет визуализировать распределение числовой переменной и сравнить его между категориями.

```python
df.boxplot(by='floors', column='price')
plt.show()
```
